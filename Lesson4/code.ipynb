## Задание 2.1

Сума вероятностей противоположных событий равна 1. Подбросим монетку 1000 раз. Вероятность выбросить орёл 500/1000, вероятность решка 500/1000

```python
n = 1000
a, r = 0, 0
for i in range(1000):
    if np.random.uniform(0, 2) > 1:
        a += 1
    else:
        r += 1

print("Сумма вероятностей: %s " % ((a+r)/n))
```

## Задание 5

```python
n = 100
r = 0.7
x = np.random.rand(n)
y = r*x + (1-r)*np.random.rand(n)
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

a = (np.sum(x)*np.sum(y) - n*np.sum(x*y))/(np.sum(x)*np.sum(x) - n*np.sum(x*x))
b = (np.sum(y) - a*np.sum(x))/n

A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y)[0]

print(a, b)
print(a1, b1)

plt.plot([0,1], [b, a+b])

xM = x.mean()
yM = y.mean()
c = np.sum((x - xM)*(y - yM))/np.sqrt(np.sum((x - xM)*(x - xM))*np.sum((y - yM)*(y - yM)))

c1 = np.corrcoef(x, y)
print('Ручной расчёт: ', c)
print('Методама Numpy: ', c1[0,1])
```
