```python
x = np.linspace(-2, 5, 201)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, np.ones(201))
plt.grid(True)
plt.ylim(-3)
```

```python
from scipy.optimize import fsolve
```

```python
def equations(args):
    x, y = args
    return (x**2 - 1 - y, np.exp(x) + x*(1-y) - 1)

x0, y0 = fsolve(equations, (-2, -1))
x1, y1 = fsolve(equations, (2, 3))
x2, y2 = fsolve(equations, (4, 5))

print(x0, y0)
print(x1, y1)
print(x2, y2)
```
