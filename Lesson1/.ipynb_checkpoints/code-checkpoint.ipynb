### Задание 3

#### Окружность

```python
R = 5
x = np.linspace(-R, R, 10000)
y1 = []
y2 = []

for X in x:
    y1.append(np.sqrt(R**2 - X**2))
    y2.append(-np.sqrt(R**2 - X**2))
    
plt.plot(x, y1)
plt.plot(x, y2)
plt.gca().set_aspect('equal', adjustable='box')
```

#### Эллипс

```python
R1 = 9
R2 = 5
x = np.linspace(-R1, R1, 10000)
y1 = []
y2 = []

for X in x:
    y1.append(np.sqrt(R2**2 - R2**2 * X**2 / R1**2))
    y2.append(-np.sqrt(R2**2 - R2**2 * X**2 / R1**2))
    
plt.plot(x, y1)
plt.plot(x, y2)
plt.gca().set_aspect('equal', adjustable='box')
```

#### Гипербола

```python
R1 = 500
R2 = 900
x1 = []
y1 = []
y2 = []

for i in range(-1000, 1000):
    X = i
    x1.append(X)

    y1.append(R2/R1 * np.sqrt(X**2 - R1**2))
    y2.append(-R2/R1 * np.sqrt(X**2 - R1**2))
    
plt.plot(x1, y1)
plt.plot(x1, y2)
plt.gca().set_aspect('equal', adjustable='box')
```

### Задание 5

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,1)
Y = np.arange(-5,5,1)
X, Y = np.meshgrid(X, Y)
Z1 = X + 2*Y
Z2 = Z1+7
ax.plot_surface(X,Y,Z1,color='red')
ax.plot_wireframe(X,Y,Z2)
```

#### Эллиптический пораболоид

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.25)
Y = np.arange(-5,5,0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2/5 + Y**2/4
ax.plot_surface(X,Y,Z)
```

#### Двухполосный гиперболоид

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-10,10,0.25)
Y = np.arange(-10,10,0.25)
X, Y = np.meshgrid(X, Y)
Z1 = np.sqrt(9*X**2/5 + 9*Y**2/3 + 9)
Z2 = -1*Z1
ax.plot_surface(X,Y,Z1)
ax.plot_surface(X,Y,Z2)
```
